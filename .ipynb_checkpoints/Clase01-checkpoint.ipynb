```python
def primeNumbers(N):
    primes = []
    for i in range(2, N+1):
        isPrime = True
        for j in range(2,i):
            if i % j == 0:
                isPrime = False
                break;
        if isPrime:
            primes.append(i)
    return primes
```

```python
primes = primeNumbers(15)
print(primes)
```

```python
def crossProduct(a,b):
    #a = (ax,au,az)
    #b = (bx,by,bz)
    product = [a[1]*b[2] - a[2]*b[1], a[2]*b[0] - a[0]*b[2], a[0] * b[1] + a[1] * b[0]]
    return product
```

```python
vectorA = [1,2,3]
vectorB = [4,5,6]
productoCruz = crossProduct(vectorA,vectorB)
print(productoCruz)
```
